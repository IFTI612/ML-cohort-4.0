MATPLOTLIB

```python
import matplotlib.pyplot as plt  #matplotlib.pyplot hocche matplotlib er sub module
import pandas as pd
import numpy as np
```

Input the data using pandas

```python
dhaka_housing = pd.read_csv("House_Rent_Dataset.csv")
```

```python
dhaka_housing.Size.plot(kind='line',color='g',lable = 'area',linewidth = 1 , alpha = 0.5 , grid=True,linestyle=':')
dhaka_housing.Rent.plot(color='r',lable = 'price',linewidth = 1 , alpha = 0.5 , grid=True,linestyle='-.')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Plot')
plt.show()
```
